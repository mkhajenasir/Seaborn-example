## The Titanic Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.pairplot(titanic, vars=["fare", "age"]);
```

```python
sns.jointplot(x= 'fare', y= 'age', data= titanic);
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(15,8))
sns.distplot(titanic['fare'], ax=axs[0]);
sns.distplot(titanic['fare'], bins=15, kde= False, color= 'green', ax=axs[1]);
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(15,8))
sns.boxplot(data=titanic, x= 'embark_town', y= 'age', ax=axs[0]);
sns.boxplot(x='class', y='age', data=titanic, palette='rainbow', ax=axs[1]);
sns.swarmplot(x='class', y='age', data=titanic, palette='Set2', ax=axs[2]);
```

```python
fig, axs = plt.subplots(ncols=1, figsize=(8,8))
sns.countplot(x='sex',data=titanic);
```

```python
g = sns.FacetGrid(titanic, col="sex", hue="class", height=8)
g.map(sns.scatterplot, "fare", "age", alpha=.7)
g.add_legend();
```

```python
fig, axs = plt.subplots(ncols=1, figsize=(8,8))
sns.heatmap(titanic.corr(),cmap='coolwarm');
plt.title('titanic correlation');
```

```python
g = sns.FacetGrid(data=titanic,col='sex', height=8);
g.map(plt.hist,'age');
```
